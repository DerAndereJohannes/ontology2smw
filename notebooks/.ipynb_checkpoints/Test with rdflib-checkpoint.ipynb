```python
import rdflib


def test_ontology_parse():
    graph = rdflib.Graph()
    graph.parse(
        source='https://raw.githubusercontent.com/tibonto/aeon/confIDent/aeon.ttl',
        format='ttl')
    assert graph

test_ontology_parse()
```

```python

# open a graph
graph = rdflib.Graph()

# load some data
graph.parse('aeon/aeon.ttl', format="ttl")  

qres = graph.query(
    """
    PREFIX aeon: <https://github.com/tibonto/aeon#>
    PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
    SELECT ?subject ?x 
    WHERE {
         ?subject aeon:SMW_datatype ?x.
    }""")
     

print("the graph has %s statements" % len(qres))

for printout in qres:
    printout_dict = printout.asdict()
    print(printout_dict)
```
